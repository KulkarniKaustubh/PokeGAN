```python
import os
import cv2
import pickle
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

DATA_PATH = 'dataset'

images = []

for pokemon in sorted(os.listdir(DATA_PATH)):
    for image in sorted(os.listdir(os.path.join(DATA_PATH, pokemon))):
        img = cv2.imread(os.path.join(DATA_PATH, pokemon, image))
        images.append(np.array(img))
        print("pokemon: " + str(pokemon))
        print(image)
        break
    break
```

```python
print(len(images))
print(type(images[0]))
print(images[0].shape)
plt.imshow(images[0])
```

```python
file = open('images.pkl', 'wb')
pickle.dump(images, file)
file.close()
```

```python
file = open('images.pkl', 'rb')
imgs = pickle.load(file)
file.close()
plt.imshow(imgs[0])
```
